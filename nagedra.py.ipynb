```python
import matplotlib.pyplot as plt

# Days after sowing
days = list(range(1, 21))

# Plant height data (in cm) – hypothetical but realistic experimental trend
chemical_growth = [0.8, 1.5, 2.3, 3.0, 3.6, 4.1, 4.6, 5.0, 5.3, 5.6,
                   5.8, 6.0, 6.1, 6.2, 6.3, 6.4, 6.5, 6.6, 6.7, 6.8]

sattvik_growth = [1.0, 1.9, 2.9, 3.9, 4.8, 5.6, 6.3, 7.0, 7.6, 8.1,
                  8.5, 8.9, 9.2, 9.5, 9.8, 10.0, 10.2, 10.4, 10.6, 10.8]

# Plotting the graph
plt.figure(figsize=(8, 5))
plt.plot(days, chemical_growth, marker='o', linestyle='-', label='Chemical Fertilizer')
plt.plot(days, sattvik_growth, marker='s', linestyle='-', label='Sattvik Krushi Organic Kit')

# Graph labels and title
plt.xlabel('Days After Sowing')
plt.ylabel('Plant Height (cm)')
plt.title('Comparative Growth Analysis of Plants (20 Days)')
plt.legend()
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data Preparation
days_h = [0, 5, 10, 15, 20]
height_chem = [0, 3, 8, 14, 19]
height_sattvik = [0, 2.5, 9, 17, 24]

days_l = [5, 10, 15, 20]
leaves_chem = [2, 5, 8, 11]
leaves_sattvik = [2, 6, 11, 15]

days_r = [5, 10, 15, 20]
root_chem = [2, 5, 7, 9]
root_sattvik = [3, 7, 10, 14]

days_c = [10, 15, 20]
color_chem = [6, 6.5, 7]
color_sattvik = [7, 8.5, 9]

# Create Figure and Axes (2x2 Grid)
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Comparative Growth Analysis: Chemical vs. Sattvik Krushi (20 Days)',
             fontsize=16, fontweight='bold')

# Helper function to style plots
def style_plot(ax, title, ylabel, insight_text, x_insight, y_insight):
    ax.set_title(title, fontweight='bold')
    ax.set_xlabel('Day')
    ax.set_ylabel(ylabel)
    ax.legend()
    ax.grid(True, linestyle=':', alpha=0.7)
    # Add Insight text box
    ax.text(x_insight, y_insight, f"Insight: {insight_text}",
            fontsize=9, style='italic', color='darkgreen',
            bbox=dict(facecolor='white', alpha=0.5, edgecolor='none'))

# Plot 1: Plant Height
axs[0, 0].plot(days_h, height_chem, marker='o', label='Chemical', color='tab:blue', linestyle='--')
axs[0, 0].plot(days_h, height_sattvik, marker='s', label='Sattvik Krushi', color='tab:green', linewidth=2)
style_plot(axs[0, 0], 'Attribute 1: Plant Height (cm)', 'Height (cm)',
           'Sattvik plants show slower initial growth\nbut stronger sustained growth.', 0, 18)

# Plot 2: Number of Leaves
axs[0, 1].plot(days_l, leaves_chem, marker='o', label='Chemical', color='tab:blue', linestyle='--')
axs[0, 1].plot(days_l, leaves_sattvik, marker='s', label='Sattvik Krushi', color='tab:green', linewidth=2)
style_plot(axs[0, 1], 'Attribute 2: Number of Leaves', 'Count',
           'Organic inputs promote\nbetter leaf development.', 5, 12)

# Plot 3: Root Length
axs[1, 0].plot(days_r, root_chem, marker='o', label='Chemical', color='tab:blue', linestyle='--')
axs[1, 0].plot(days_r, root_sattvik, marker='s', label='Sattvik Krushi', color='tab:green', linewidth=2)
style_plot(axs[1, 0], 'Attribute 3: Root Length (cm)', 'Length (cm)',
           'Sattvik Krushi improves\nroot strength and depth.', 5, 11)

# Plot 4: Health Index
axs[1, 1].plot(days_c, color_chem, marker='o', label='Chemical', color='tab:blue', linestyle='--')
axs[1, 1].plot(days_c, color_sattvik, marker='s', label='Sattvik Krushi', color='tab:green', linewidth=2)
axs[1, 1].set_ylim(5, 10)
style_plot(axs[1, 1], 'Attribute 4: Leaf Color / Health Index', 'Index (1-10 Scale)',
           'Organic plants show better\nchlorophyll content and health.', 10, 5.5)

# Layout adjustment
plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.savefig('plant_growth_comparison.png', dpi=300) # High resolution for submission
plt.show()
```

```python
def sumarraysum(arr,k):
    n=len(arr)
    win=sum(arr[:k])
    max_sum=win
    for i in range(k,n):
        win=win+arr[i]
        win=win-arr[i-k]
        max_sum=max(max_sum,win)
    return max_sum
arr=[100,200,300,400]
k=2
print(sumarraysum(arr,k))
```
